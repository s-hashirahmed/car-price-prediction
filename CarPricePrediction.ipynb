```python
def get_df(name, trans, eng_type):
    x = df[(df['Name']==name)]
    x = x[x['Transmission'] == trans]
    x = x[x['Engine Type'] == eng_type]
    
    return x[['Model Year', 'Mileage', 'Engine Capacity']], x['Price']

name = 'Toyota Corolla GLi Automatic 1.3 V'
trans = 'Automatic'
eng_type = 'Petrol'

get_df(name, trans, eng_type)
```

```python
from sklearn.model_selection import train_test_split
x, y = get_df(name, trans, eng_type)

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
# from sklearn.Linear_model import Linear
from sklearn.linear_model import LinearRegression
model_1 = DecisionTreeRegressor()
model_1.fit(X_train,y_train)

model_2 = SVR()
model_2.fit(X_train, y_train)

model_3 = LinearRegression().fit(X_train, y_train)
```
